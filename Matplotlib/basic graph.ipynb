##### Load Lib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Graph

```python
x = [1, 2, 3]
y = [11, 22, 33]
plt.title('Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(x, y)
```

```python
x = [1, 2, 3]
y = [11, 22, 33]
plt.title('Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
"""
自定義刻度
"""
plt.xticks([0, 1, 2])
plt.yticks([11, 22, 33, 44])
plt.plot(x, y)
```

```python
x = [1, 2, 3]
y = [11, 22, 33]
plt.title('Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0, 1, 2])
plt.yticks([11, 22, 33, 44])
"""
color 定義線條顏色
linewidth 定義線條大小
linestyle 定義線條樣式
marker 標記數據的樣式
markersize marker 大小
markeredgecolor marker 顏色
color、marker、linestyle 可以用 'b^--' 表示
"""
plt.plot(x, y, label='11x', color='r', linewidth=6, linestyle='--', marker='x', markersize=10, markeredgecolor='b')
"""
顯示 plot 定義的 label
"""
plt.legend()
```

```python
x = [1, 2, 3]
y = [11, 22, 33]

x2 = np.arange(0, 5.5, 0.3)
plt.plot(x2, x2**2, 'b', label='x^2')

plt.title('Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0, 1, 2, 3, 4, 5])
plt.yticks([11, 22, 33, 44])
"""
color 定義線條顏色
linewidth 定義線條大小
linestyle 定義線條樣式
marker 標記數據的樣式
markersize marker 大小
markeredgecolor marker 顏色
color、marker、linestyle 可以用 'b^--' 表示
"""
plt.plot(x, y, 'r^--', label='11x', linewidth=6, markersize=10, markeredgecolor='b')
"""
顯示 plot 定義的 label
"""
plt.legend()
"""
定義圖片顯示大小
dpi 像素
"""
plt.figure(figsize=(5, 3), dpi=400)
"""
儲存
"""
plt.savefig('graph.png', dpi=500)
```

## Bar chart

```python
labels=['A', 'B', 'C']
values = [2, 4, 3]
plt.bar(labels, values)
```

```python
labels=['A', 'B', 'C']
values = [2, 4, 3]
bars = plt.bar(labels, values)
"""
標註線條
"""
bars[0].set_hatch('X')
bars[1].set_hatch('/') 
bars[2].set_hatch('*')
```
