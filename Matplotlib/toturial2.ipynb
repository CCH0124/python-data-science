```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Line Grapg

```python
gas = pd.read_csv('Matplotlib\\gas_prices.csv')
plt.figure(figsize=(8, 5))

plt.plot(gas.Year, gas.USA, 'r.--', label="USA")
plt.plot(gas.Year, gas.Canada, 'b.--', label="Canada")
plt.plot(gas.Year, gas['South Korea'], 'y.--', label="SK")

plt.title('Gas Prices')
plt.ylabel('Dollars/Gallon')
plt.legend()

plt.xticks(gas.Year[::3].tolist()+[2011]) # 
plt.show()
```

```python
gas = pd.read_csv('Matplotlib\\gas_prices.csv')
plt.figure(figsize=(10, 10), dpi=100)

for country in gas.columns[1:]:
    plt.plot(gas.Year, gas[country], marker='.')

plt.title('Gas Prices')
plt.ylabel('Dollars/Gallon')
plt.legend()

plt.xticks(gas.Year[::3].tolist()+[2011]) # 
plt.show()
```

## Histogram

```python
fifa = pd.read_csv('Matplotlib\\fifa_data.csv')
```

```python
plt.figure(figsize=(8, 5), dpi=100)
a = plt.hist(fifa.Overall, bins=10, color='#aabbdd')

plt.ylabel('Number of players')
plt.xlabel('Fifa Skill Level (Overall)')
plt.show()
```

## Pie chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.pie([left, right], labels=['left', 'right'], colors=['#abcdef', '#abccff'], autopct='%2.f %%', explode=[.1, .2])
plt.title('FIFA ')
plt.show()
```

```python
plt.style.use('default')
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
labels = ['FC Barcelona', 'Real Madrid']
plt.boxplot([barcelona, madrid], labels=labels)
# use 'set' setting linewidth or color
plt.ylabel('FIFA Overall')
plt.show()
```
